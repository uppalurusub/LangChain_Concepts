#Document Transformers/ Text Splitters

LangChain offers different text splitters for splitting the data such as:

* Split by Character
* Recursive Splitter
* Token Splitter

**Split by Character** - This is the simplest method. This splits based on characters (by default "\n\n") and measure chunk length by number of characters.
* How the text is split: by single character.
* How the chunk size is measured: by number of characters.

**Recursive Text Splitter** - This text splitter is the recommended one for generic text. It is parameterized by a list of characters. It tries to split on them in order until the chunks are small enough. The default list is ["\n\n", "\n", " ", ""]. This has the effect of trying to keep all paragraphs (and then sentences, and then words) together as long as possible, as those would generically seem to be the strongest semantically related pieces of text.
* How the text is split: by list of characters.
* How the chunk size is measured: by number of characters.

**Split by tokens** - Language models have a token limit. You should not exceed the token limit. When you split your text into chunks it is therefore a good idea to count the number of tokens. There are many tokenizers. When you count tokens in your text you should use the same tokenizer as used in the language model.

##CharacterTextSplitter

**Concept**


Splits text based on a specific character (e.g., a space, newline, or period).

It attempts to break text into chunks of a fixed size, ensuring they do not exceed a maximum chunk size.

If the exact split point is not found within the desired chunk size, it still breaks at the next available occurrence of the character.

**Use Cases**


When working with structured documents where clear delimiters (e.g., newlines, punctuation) exist.

Good for breaking paragraphs or structured text files like JSON, CSV, or Markdown.

**Limitations**


If the chosen delimiter does not appear frequently, chunks might be too long.

If the delimiter appears too often, chunks might be too short.

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
splitter = CharacterTextSplitter(
    separator="\n",  # Split by newlines
    chunk_size=50,
    chunk_overlap=10
)
```

```python
import warnings
warnings.filterwarnings("ignore")

chunks = splitter.split_text(text)
```

##RecursiveCharacterTextSplitter

**Concept**


Works recursively, meaning it tries multiple split strategies.

Starts with a primary character, e.g., \n\n (new paragraph), then falls back to \n (new line), then . (sentence), etc.

Ensures that chunks remain meaningful while fitting within the max chunk size.

**Use Cases**


When text has an unknown or inconsistent structure (e.g., free-form text, PDFs, scraped web content).

When trying to keep semantic meaning intact, ensuring chunks are not cut mid-sentence.

When dealing with large, unstructured documents like Wikipedia pages, research papers, or chat transcripts.

Why Use RecursiveCharacterTextSplitter?


✅ Flexible – Adapts to different types of text.


✅ Intelligent Splitting – Preserves semantic meaning better than naive splitting.


✅ Efficient – Works well with document retrieval, embeddings, and LLM context limits.

```python
# Import RecursiveCharacterTextSplitter
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=10,
    separators=["\n\n", "\n", ". ", " "]  # Order of splitting
)
```

```python
chunks = splitter.split_text(text)
```

```python
print("Printing chunks \n\n")
for i in range(len(chunks)):
    print(i,chunks[i])
```

```python
print("Chunk 1: ", chunks[0], "\n")
print("Chunk 2: ", chunks[1], "\n")
print("Chunk 3: ", chunks[2], "\n")
```

#Text Embeddings

The Embeddings class is a class designed for interfacing with text embedding models. LangChain provides support for most of the embedding model providers (OpenAI, Cohere) including sentence transformers library from Hugging Face.

Embeddings create a vector representation of a piece of text and supports all the operations such as similarity search, text comparison, sentiment analysis etc.

The base Embeddings class in LangChain provides two methods: one for embedding documents and one for embedding a query.

The first method takes as input multiple texts, while the second method returns the embedding representation for a single text.


##CharacterTextSplitter

```python
from langchain.text_splitter import CharacterTextSplitter

# Sample long text
text = """LangChain simplifies AI application development by providing robust tools
for integrating LLMs with memory, retrieval, and reasoning mechanisms."""

# Initialize Text Splitter
text_splitter = CharacterTextSplitter(chunk_size=50, chunk_overlap=10)

# Split Text
chunks = text_splitter.split_text(text)

# Output chunks
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk}")
```

```python
# Initialize OpenAI Embedding Model
embedding_model = OpenAIEmbeddings()

# Generate embeddings for each chunk
embeddings = [embedding_model.embed_query(chunk) for chunk in chunks]

print(f"Generated {len(embeddings)} embeddings.")
```
