# Model I/O


LangChain's Model I/O component provides support to interface with the LLM and generate responses.

The Model I/O consists of:
* **Prompts**: Templatize, dynamically select, and manage model inputs
* **Language Models**: Make calls to language models through common interfaces
* **Output Parsers**: Extract information from model outputs

The general flow of Model I/O in LangChain is illustrated in the image below.

image.png

**Model I/O**


LangChain provides an easy out-of-the box support to work with LLMs. LangChain provides interfaces and integrations for two classes of LLM models


*   **LLMs**: Models that take a text string as input and return a text string


*   **Chat models**: Models that are backed by a language model but take a list of Chat Messages as input and return a Chat Message.

#Difference between LLMs and Chat Models

- **Large Language Models (LLMs):**<br><br>

LLMs are like super-smart text generators. They're great at understanding language and can write coherent and relevant text for a wide range of tasks. For example, they can summarize articles, translate languages, or even write stories.

**Focus:**

They focus on generating text.<br><br>
**How They Work:**<br><br>
LLMs learn from huge amounts of text data to understand language patterns.<br><br>
**Strengths:**<br><br>
They're flexible and powerful at understanding and creating text.<br><br>
Examples: GPT-3 is a famous example of a large language model.<br><br>








- **Chat Models:**

Chat Models are like conversational partners. <br><br>

They're specifically designed to have interactive conversations with people. They aim to understand what you're saying and respond in a way that makes sense in a conversation.

**Focus:** <br><br>

They're all about interactive conversations.<br><br>
**How They Work:**<br><br>

Chat Models pay attention to both what's said before and after in a conversation to understand context better.<br><br>
**Strengths:**<br><br>

They're good at keeping track of what's been said before and making conversation feel natural.<br><br>
Examples: OpenAI's ChatGPT is a well-known chat model.

#Prompt Serialization

Prompts can be saved or 'serialized' instead of being written multiple times.

**What is Prompt Serialization?**
Prompt Serialization refers to the process of structuring and formatting prompts in a way that they can be consistently used and understood by machines. This involves transforming a prompt into a serialized format that can be easily stored, transmitted, and processed by software systems.

**Key Points:**

Serialization is the process of converting an object into a format that can be easily stored or transmitted.
In the context of prompts, serialization ensures that the prompts maintain their structure and meaning when used in different environments or applications.

Why is Prompt Serialization Important?
Prompt serialization is crucial for several reasons:
<br>
**Consistency**: Ensures that prompts maintain their intended structure and meaning across different platforms and applications.
<br>
**Interoperability**: Facilitates the use of prompts in various systems by adhering to standardized formats.
Storage and Transmission: Allows prompts to be stored in databases or transmitted over networks efficiently.
<br>
**Reusability**: Enables prompts to be reused in different contexts without losing their integrity.<br>
**Applications of Prompt Serialization**
Prompt serialization is widely used in the following areas:

**Natural Language Processing (NLP)**: Ensures that language models can understand and process prompts accurately.<br>
**Chatbots and Virtual Assistants**: Helps in maintaining consistent communication patterns and responses.<br>
**Data Interchange**: Facilitates the exchange of prompts between different systems and applications.
Machine Learning Models: Assists in training and deploying models with consistent input data.

#Few Shot Prompt Template

```python
# Import the FewShot Prompt Template
from langchain.prompts import FewShotPromptTemplate

# create a example template
example_template = """
User: {query}
AI: {answer}
"""

# create our examples
examples = [
    {
        "query": "How are you?",
        "answer": "I can't complain but sometimes I still do."
    }, {
        "query": "What time is it?",
        "answer": "It's time to get a watch."
    }
]

# create a prompt example from above template
example_prompt = PromptTemplate(
    input_variables = ["query", "answer"],
    template = example_template
)

# now break our previous prompt into a prefix and suffix
# the prefix is our instructions
prefix = """The following are exerpts from conversations with an AI
assistant. The assistant is typically sarcastic and witty, producing
creative  and funny responses to the users questions. Here are some
examples:
"""

# and the suffix our user input and output indicator
suffix = """
User: {query}
AI: """

# now create the few shot prompt template
few_shot_prompt_template = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix=prefix,
    suffix=suffix,
    input_variables=["query"],
    example_separator="\n\n"
)
```

```python
query = "What is the meaning of life?"
```

```python
few_shot_request = few_shot_prompt_template.format(query=query)
```

```python
print(few_shot_request)
```

```python
few_shot_response = llmModel.invoke(few_shot_request)
```

#Chains

In LangChain, a **"chain"** refers to a sequence of actions or steps that are linked together to achieve a specific goal. LangChain is a framework designed to facilitate the creation of applications that integrate large language models (LLMs) and other AI tools in a structured and effective manner. Chains are central to how LangChain orchestrates complex workflows involving these models.

Types of Chains:

**Simple Chains:**
These are linear sequences of operations where each step depends on the output of the previous step.
For example, a chain might first call an LLM to generate text, then process this text with a custom function, and finally store the result in a database.

**Multi-step Chains:**
These involve more complex workflows where multiple steps are executed in a predefined order.
They can include conditional logic, loops, and parallel processing to handle more intricate tasks.

**Custom Chains:**
Users can define their own chains by specifying the sequence of steps and the logic for how they interact.
Custom chains can integrate various tools and services, such as other APIs, databases, and external services.

Using an LLM in isolation is fine for simple applications, but more complex applications require chaining LLMs - either with each other or with other components.

Example: <br>
Here’s a simple example of a chain in LangChain:

**Step 1:** Call a language model to generate an initial draft of a blog post based on a given topic.<br>
**Step 2:** Use a custom function to analyze the draft and identify key points that need further elaboration.<br>
**Step 3:** Call the language model again to expand on those key points.<br>
**Step 4:** Perform grammar and style checking on the expanded text.<br>
**Step 5:** Store the final text in a content management system.<br>

By chaining these steps together, you create an automated workflow that generates high-quality content with minimal manual intervention.

Chains in LangChain are powerful constructs that enable the integration and coordination of multiple AI-driven steps into cohesive workflows, simplifying the development of complex applications that leverage large language models.

##Router Chain

**What it is:** A chain that routes requests to different chains based on input.

**How it works:** It decides which sub-chain should handle the request.

**Example Use Case:** Directing different types of user queries (e.g., FAQs vs. complex reasoning).

**When to Use:** When dynamically selecting chains based on user queries.

In LangChain, a "Router Chain" refers to a type of chain designed to route or direct tasks to different sub-chains or components based on certain conditions or criteria. Router Chains are particularly useful for building complex workflows where different types of inputs need to be processed by different logic or components.


Router Chains involves creating a chain that dynamically selects the next chain to use for a given input.

LLMRouterChains takes in an input and redirect it to the most appropriate LLMChain sequence.

The LLMRouterChain accepts multple potential destination LLMChains. Then via a specialised prompt, the Router will read the initial input and then ouput a specific dictionary that matches one of the potential destination chains to continue the processing.

Router chains are made up of two components:
* The RouterChain itself (responsible for selecting the next chain to call)
* destination_chains: chains that the router chain can route to



###Example 1

```python
from langchain.chains.router import MultiPromptChain
from langchain.chains import ConversationChain
```

##Conversation Buffer Memory


ConversationBufferMemory keeps a list of chat messages in a buffer and passes those into the prompt template.

**Concept:**


It stores the entire conversation history in a buffer.

When a user interacts with the AI, the model refers to the full conversation context.

**Use Cases:**


Simple chatbot applications where retaining all previous interactions is needed.

Customer service bots tracking a full conversation with a user.

Assistants for research where all details need to be remembered.

**Limitations:**


As conversation history grows, memory can become too large, increasing processing costs.

Not efficient for long interactions.

#Pydantic to define a custom output format

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.pydantic_v1 import Field
from pydantic import BaseModel
```

```python
# Define a Pydantic Object with the desired output format.
class Joke(BaseModel):
    setup: str = Field(description="question to set up a joke")
    punchline: str = Field(description="answer to resolve the joke")
```

```python
import warnings
warnings.filterwarnings("ignore")

# Define the parser referring the Pydantic Object
parser = JsonOutputParser(pydantic_object=Joke)

# Add the parser format instructions in the prompt definition.
prompt = PromptTemplate(
    template="Answer the user query.\n{format_instructions}\n{query}\n",
    input_variables=["query"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

# Create a chain with the prompt and the parser
chain = prompt | chatModel | parser
```

```python
response = chain.invoke({"query": "Tell me a joke."})
```
